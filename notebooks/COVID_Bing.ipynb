```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
```

# Confirmados

```
# Bing
url = 'https://raw.githubusercontent.com/microsoft/Bing-COVID-19-Data/master/data/Bing-COVID19-Data.csv'
df = pd.read_csv(url)
```

```
# Johns Hopkins
# url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
# df = pd.read_csv(url)
```

```
df['Updated'] = pd.to_datetime(df['Updated'])
df.head()
```

```
df['Country_Region'].unique()
```

```
# selecao por país e estado (AdminRegion1)
country = 'Brazil'
state = 'São Paulo'
aux = df.loc[(df['Country_Region'] == country) &
             (df['AdminRegion1'] == state)].copy()
aux.sort_values(by=['Updated'], inplace=True)
aux.set_index(['Updated'], inplace=True)
aux['Confirmed'].plot();
```

```
# selecao por país
country = 'Brazil'
aux = df.loc[(df['Country_Region'] == country) &
             (df['AdminRegion1'].isnull()) &
             (df['AdminRegion2'].isnull())].copy()
aux.sort_values(by=['Updated'], inplace=True)
aux.set_index(['Updated'], inplace=True)
aux['Confirmed'].plot();
```

```
country = 'Brazil'

fig, ax = plt.subplots(figsize=(8, 6))
aux = df.loc[(df['Country_Region'] == country) &
             (df['AdminRegion1'].isnull()) &
             (df['AdminRegion2'].isnull())].copy()
aux.sort_values(by=['Updated'], inplace=True)
aux.set_index(['Updated'], inplace=True)
aux['Confirmed'].plot(ax=ax);
aux['Deaths'].plot(ax=ax);
aux['Recovered'].plot(ax=ax);
plt.legend()
plt.grid()
plt.show()
```

```
country = 'Brazil'

fig, ax = plt.subplots(figsize=(8, 6))
aux = df.loc[(df['Country_Region'] == country) &
             (df['AdminRegion1'].isnull()) &
             (df['AdminRegion2'].isnull())].copy()
aux.sort_values(by=['Updated'], inplace=True)
aux.set_index(['Updated'], inplace=True)
aux['Confirmed'].plot(ax=ax);
aux['Deaths'].plot(ax=ax);
aux['Recovered'].plot(ax=ax);
plt.yscale('log')
plt.legend()
plt.grid()
plt.show()
```
